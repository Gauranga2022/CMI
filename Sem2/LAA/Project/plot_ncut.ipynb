
# Normalized Cut

This example constructs a Region Adjacency Graph (RAG) and recursively performs
a Normalized Cut on it [1]_.

## References
.. [1] Shi, J.; Malik, J., "Normalized cuts and image segmentation",
       Pattern Analysis and Machine Intelligence,
       IEEE Transactions on, vol. 22, no. 8, pp. 888-905, August 2000.
## Code summary
Import Necessary Libraries:

1. **skimage.data**: Provides access to sample images and example data.
2. **skimage.segmentation**: Contains image segmentation algorithms.
3. **skimage.color**: Functions for color space conversion.

Load Sample Image:
4. img = data.coffee(): Loads a sample image of coffee from skimage.data.

Segment the Image Using SLIC:

5. **labels1 = segmentation.slic(img, compactness=30, n_segments=400, start_label=1)**: Applies the SLIC algorithm to segment the image. SLIC is used to partition the image into n_segments of compact, nearly uniform superpixels. The compactness parameter balances color proximity and space proximity.

6. **out1 = color.label2rgb(labels1, img, kind='avg', bg_label=0)**: Converts the labeled image (labels1) into an RGB image where each segment's color is the average of the colors of its constituent pixels.

7. **Create a Region Adjacency Graph (RAG) and Apply Normalized Cut**:
g = graph.rag_mean_color(img, labels1, mode='similarity'): Constructs a RAG where each node represents a segment from labels1, and edges are weighted by color similarity.

8. **labels2 = graph.cut_normalized(labels1, g)**: Applies a normalized cut on the RAG. This method partitions the graph to minimize the cut cost between different segments while maximizing the similarity within the segments.
out2 = color.label2rgb(labels2, img, kind='avg', bg_label=0): Converts the new segmentation labels (labels2) into an RGB image similar to out1.

**Plotting the Results:**
9. **fig, ax = plt.subplots(nrows=2, sharex=True, sharey=True, figsize=(6, 8))**: Creates a figure with two subplots to display the images.
10. **ax[0].imshow(out1)**: Displays the image after SLIC segmentation in the first subplot.
11. **ax[1].imshow(out2)**: Displays the image after applying the normalized cut to the RAG in the second subplot.
12. Each axis is set to not display any coordinate axis using **a.axis('off')**.
13. **plt.tight_layout()**: Adjusts the layout to make sure there is no overlap in content.

```python
from skimage import data, segmentation, color
from skimage import graph
from matplotlib import pyplot as plt

img = data.coffee()

labels1 = segmentation.slic(img, compactness=30, n_segments=4000, start_label=1)
out1 = color.label2rgb(labels1, img, kind='avg', bg_label=0)

g = graph.rag_mean_color(img, labels1, mode='similarity')
labels2 = graph.cut_normalized(labels1, g)
out2 = color.label2rgb(labels2, img, kind='avg', bg_label=0)

fig, ax = plt.subplots(nrows=2, sharex=True, sharey=True, figsize=(6, 8))

ax[0].imshow(out1)
ax[1].imshow(out2)

for a in ax:
    a.axis('off')

plt.tight_layout()
```
