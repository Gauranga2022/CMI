````python
class Node:
    def __init__(self, v = None):
        self.value = v
        self.tag = "L"
        self.left = None
        self.right = None

    def find(self, v):
        if self.tag == "L":
            return self.value == v
        else:
            if self.value < v:
                return self.left.find(v)
            else:
                return self.right.find(v)

    def insert(self, v):
        if self.value == None:
            self.value = v
            return
        if self.tag == "I":
            if v > self.value:
                self.right.insert(v)
                return
            elif v < self.value:
                self.left.insert(v)
                return
            return
        else:
            if self.value == v:
                return
            if v < self.value:
                self.left = Node(v)
                self.right = Node(self.value)
                self.tag = "I"
            if v > self.value:
                self.left = Node(self.value)
                self.right = Node(v)
                self.tag = "I"
                self.value = v

    def delete(self, v):
        if self.value == None:
            return
        if self.tag == "L":
            if self.value == v:
                self.value = None
                return
            else:
                return
        if self.left.tag == "L" and self.right.tag == "L":
            if v < self.value:
                if v == self.left.value:
                    self.left = None
                    self.value = self.right.value
                    self.right = None
                    self.tag = "L"
                return
            else:
                if self.right.value == v:
                    self.value = self.left.value
                    self.left = None
                    self.right = None
                    self.tag = "L"
                return
        if self.left.tag == "L" and self.right.tag == "I":
            if v < self.value:
                if v == self.left.value:
                    self.value = self.right.value
                    self.left = self.right.left
                    self.right = self.right.right
                return
            else:
                self.right.delete(v)
                if v == self.value:
                    self.value = self.right.smallest()
                return
        if self.left.tag == "I" and self.right.tag == "L":
            if v < self.value:
                self.left.delete(v)
                return
            else:
                if v == self.value:
                    self.value = self.left.value
                    self.right = self.left.right
                    self.left = self.left.left
                return
        if self.left.tag == "I" and self.right.tag == "I":
            if v < self.value:
                self.left.delete(v)
                return
            else:
                self.right.delete(v)
                if v == self.value:
                    self.value = self.right.smallest()
                return

    def smallest(self):
        if self.left == None:
            return self.value
        else:
            return self.left.smallest()

    def inorder(self):
        if self.value == None:
            return []
        if self.left == None:
            return [(self.tag, self.value)]
        else:
            return self.left.inorder() + [(self.tag, self.value)] + self.right.inorder()

    def __str__(self):
        return str(self.inorder())
````
