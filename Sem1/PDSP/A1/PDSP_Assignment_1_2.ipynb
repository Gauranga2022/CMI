Write a function nestingdepth(s) that takes as input a string s and computes the maximum nesting depth of brackets if s has properly nested brackets. If the string is not properly matched, your function should return -1.

```python
def nestingdepth(s):
  counter,depth, max_so_far = 0,0,0
  l,r=0,0
  while counter < len(s):
    if s[counter]=='(':
      depth+=1
      l+=1
    elif s[counter]==')':
      depth-=1
      r+=1
    max_so_far = max(max_so_far,depth)
    if r>l:
      return(-1)
    counter+=1
  if depth!=0:
    return(-1)
  return(max_so_far)
```

```python
nestingdepth("((jkl)78(A)&l(8(dd(FJI:),):)?)")
```

Write a function accordian(l) that takes as input a list of integer l and returns True if the absolute difference between each adjacent pair of elements alternates between increasing strictly and decreasing strictly.

```python
def accordian(l):
  abs_list,seq,flag = [],'',True
  for i in range(len(l)-1):
    abs_list.append(abs(l[i+1]-l[i]))
  for i in range(len(abs_list)-1):
    if abs_list[i+1]<abs_list[i]:
      seq+='d'
    elif abs_list[i+1]>abs_list[i]:
      seq+='i'
    else:
      seq+='e'
  if seq.count('dd')>0 or seq.count('ii')>0 or seq.count('e')>0:
    flag = False
  return(flag)
```

```python
accordian([1,5,2,8,2])
```

Rotate a matrix by 90 degrees

```python
def rotate(mat):
  new_mat = [[0 for i in range(len(mat))]for j in range(len(mat))]
  for i in range(len(mat)):
    for j in range(len(mat)):
      new_mat[j][len(mat)-i-1]=mat[i][j]
  return(new_mat)
```

```python
rotate([[1,2],[3,4]])
```

Write a function delchar(s,c) that takes as input strings s and c, where c has length 1 (i.e., a single character), and returns the string obtained by deleting all occurrences of c in s. If c has length other than 1, the function should return s

```python
def delchar(s,c):
  arr =''
  for i in range(len(s)):
    if s[i] != c:
      arr+=s[i]
  return(arr)
```

```python
delchar('banana','a')
```
